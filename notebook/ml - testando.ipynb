```python
import pandas as pd

dados = pd.read_csv('C://fontes_dados//ml//tudo.csv')
dados = dados.query('ano == 2020')
dados = dados.query('carteira_ativa <= 100000')
#dados = dados.query('uf == "AC"')
#dados = dados.query('cliente == 1')
dados
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
dados.dropna(inplace = True)
dados.isnull().sum()

del dados['Unnamed: 0']
del dados['data_base']
del dados['uf']
del dados['ocupacao']
del dados['numero_de_operacoes']
del dados['ano']
dados
```

```python
dados.info()
```

```python
y = dados['carteira_ativa']
x = dados.drop('carteira_ativa', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

resultado = modelo.score(x_teste, y_teste)
print('Acuracia: ', resultado)
```
